```python
import pandas as pd
# load ICAP-Data
df_co2 = pd.read_csv('icap-graph-price-data-2015-01-01-2025-09-12.csv',
                     delimiter=",")  

# look at data
print(df_co2.columns)
print("\nErste 5 Zeilen:")
print(df_co2.head())

# Parse Date
df_co2['Date'] = pd.to_datetime(df_co2['Date'])

# checking March 2020
march_co2 = df_co2[(df_co2['Date'].dt.year == 2020) & (df_co2['Date'].dt.month == 3)]
print(f"\March 2020 Datapoints: {len(march_co2)}")
print(f"First: {march_co2['Date'].min()}")
print(f"Last: {march_co2['Date'].max()}")

# should be: 
# - about 22 Datapoints (Working days in March)
# # - last: 2020-03-31 (Trading Date 31. march)
```
