```python
import pandas as pd
# Ignoriere Problem-Timestamps für diesen Test
df_raw = pd.read_csv('Day Ahead Preise_M15_2020_English.csv')
df_raw['timestamp'] = pd.to_datetime(df_raw['Time from [CET/CEST]'], errors='coerce')

# Filtere NaT (die 2A/2B Zeilen) raus für den Test
df_raw_clean = df_raw[df_raw['timestamp'].notna()]

# Letzter Datenpunkt März 2020
last_march = df_raw_clean[df_raw_clean['timestamp'].dt.month == 3].iloc[-1]
print(f"Letzter März-Timestamp: {last_march['timestamp']}")

# Zähle März-Datenpunkte (ohne die wenigen NaT)
march_count = len(df_raw_clean[df_raw_clean['timestamp'].dt.month == 3])
expected = 31 * 96  # 2976
print(f"März-Datenpunkte: {march_count} (erwartet: {expected})")

# Wenn ~2976 → Delivery Date ✓
# Wenn ~2880 → Trading Date Problem
```
